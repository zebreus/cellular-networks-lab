```typescript
import * as Plot from "npm:@observablehq/plot";
import { DOMParser, SVGElement } from "npm:linkedom";
import { display } from "https://deno.land/x/display/mod.ts";

const document = new DOMParser().parseFromString(
  `<!DOCTYPE html><html lang="en"></html>`,
  "text/html",
);

const car_json = await fetch("https://raw.githubusercontent.com/vega/vega/main/docs/data/cars.json")
const cars = await car_json.json()
```

```typescript
const { display, md,svg } = Deno.jupyter;
let svgElement = await Plot.ruleX(cars, {x: "Weight_in_lbs"}).plot({document})
 

await display(svg`${svgElement.toString()}`)

"sdfa"
```
